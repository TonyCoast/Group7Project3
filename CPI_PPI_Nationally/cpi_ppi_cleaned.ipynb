```python
import pandas as pd

url = "/Users/tanneramman/Downloads/Merged_PPI_Data.csv"
url2 = "/Users/tanneramman/Downloads/Merged_CPI_Data.csv"
ppi = pd.read_csv(url)
cpi = pd.read_csv(url2)

cpi.head()
```

```python
print(cpi.dtypes)
```

```python
cpi.columns
```

```python
cpi['month'] = cpi['period'].str.extract('(\d+)').astype(int)
cpi.head()
```

```python
ppi['month'] = ppi['period'].str.extract('(\d+)').astype(int)
ppi.head()
```

```python
cpi['Date'] = pd.to_datetime(cpi[['year', 'month']].assign(DAY=1))
cpi.head()
```

```python
ppi['Date'] = pd.to_datetime(ppi[['year', 'month']].assign(DAY=1))
ppi.head()
```

```python
cpi = cpi.drop('period', axis=1)
cpi.head()
```

```python
cpi = cpi.drop('year', axis=1)
cpi = cpi.drop('month', axis=1)
cpi.head()
```

```python
ppi = ppi.drop('period', axis=1)
ppi = ppi.drop('year', axis=1)
ppi = ppi.drop('month', axis=1)
ppi.head()
```

```python
cpi.to_csv("/Users/tanneramman/Downloads/cpi.csv", index=False)

ppi.to_csv("/Users/tanneramman/Downloads/ppi.csv", index=False)
```

```python
cpi = cpi.rename(columns={'Dairy': 'DairyCPI', 'Alcohol': 'AlcoholCPI', 'Meats': 'MeatsCPI'}) 
cpi.head()
```

```python
ppi = ppi.rename(columns={'Dairy ': 'DairyPPI', 'Alcohol': 'AlcoholPPI', 'Meats': 'MeatsPPI'}) 
ppi.head()
```

```python
cpi_ppi_merged = pd.merge(cpi, ppi)
cpi_ppi_merged.head()
```

```python
cpi_ppi_merged.columns
```

```python
cpi_ppi_merged = cpi_ppi_merged[['Date', 'DairyCPI', 'AlcoholCPI', 'MeatsCPI', 'DairyPPI', 'AlcoholPPI', 'MeatsPPI']]
cpi_ppi_merged.head()
```
